```python
df = (df.groupby(['season', 'Team', 'win_match', 'position'], dropna=False)['overall_pick'].sum()/df.groupby(['season', 'Team', 'win_match', 'position'], dropna=False)['overall_pick'].count()).reset_index(name='total_pick')
```

```python
df.isnull().sum()
```

```python
Max = 61
Min = 1
df['total_pick'] = df.apply(lambda x: Max + Min - x['total_pick'], axis=1)
```

```python
def category(valeur):
    if valeur == "nan":
        pass
    elif valeur <= 10:
        return 'A'
    elif valeur <= 20:
        return "B"
    elif valeur <= 30:
        return "C"
    elif valeur <= 40:
        return "D"
    elif valeur <= 50:
        return "E"
    elif valeur <= 60:
        return "F"
    elif valeur <= 70:
        return "G"
    elif valeur <= 80:
        return "H"
    else:
        return "I"
    
```

```python
df['win_match'] = df['win_match'].apply(category)
df
```
