```python
df_game = pd.read_csv('game.csv')
df_game.head()
```

```python
df_game['game_date'] = pd.to_datetime(df_game['game_date'])
df_game['year'] = df_game['game_date'].dt.year.astype(int)
df_game= df_game[df_game['year'] >= 2010]
df_game = df_game[df_game['season_type'].isin(['Regular Season'])]
```

```python
df_game.columns
```

```python
new_game = df_game.loc[:,['season_id', 'game_id', 'team_name_home', 'wl_home', 'team_name_away', 'wl_away']]
new_game
```

```python
game_by_team_home = new_game.groupby(['team_name_home', 'season_id'])['wl_home'].value_counts(ascending=True).reset_index(name='total')
game_by_team_home
```

```python
game_by_team_away = new_game.groupby(['team_name_away', 'season_id'])['wl_away'].value_counts(ascending=True).reset_index(name='total')
game_by_team_away.sort_values(by=['team_name_away', 'season_id','wl_away'])
```

```python
research = new_game.groupby(['team_name_away', 'season_id'])['wl_away'].value_counts(ascending=True).reset_index(name='total_away')
amb1_1 = new_game.groupby(['team_name_home', 'season_id'])['wl_home'].value_counts(ascending=False).reset_index(name='total_home')
research.insert(loc=len(research.columns), column='total_home', value=amb1_1['total_home'])
colomns_list = ['total_home', 'total_away']
research['somme'] = research[colomns_list].sum(axis=1)
research
```

```python
research = new_game.groupby(['team_name_away', 'season_id'])['wl_away'].value_counts().reset_index(name='total_away')
amb1_1 = new_game.groupby(['team_name_home', 'season_id'])['wl_home'].value_counts().reset_index(name='total_home')
research.insert(loc=len(research.columns), column='total_home', value=amb1_1['total_home'])
research = research.sort_values(by=['team_name_away', 'season_id', 'wl_away'])
colomns_list = ['total_home', 'total_away']
research['somme'] = research[colomns_list].sum(axis=1)
research
```

```python
new = research.loc[research['wl_away'] == 'W']
new
```

```python
def supprimer_premier_2(valeur):
    valeur_str = str(valeur)
    return int(valeur_str.replace('2', '', 1)) if '2' in valeur_str else valeur
```

```python
new_game2 = new.loc[:,['team_name_away', 'season_id', 'somme']]
new_game2['season_id'] = new_game2['season_id'].apply(supprimer_premier_2)
new_game2
```

```python
new_game2['season'] = pd.to_datetime(new_game2['season_id'], format='%Y').dt.year
new_game2
```

```python
new_game2 = new_game2.loc[:,['team_name_away', 'season', 'somme']]
new_game2
```

```python
import matplotlib.pyplot as plt
```

```python

# Tracer un graphique en ligne pour chaque nom
plt.figure(figsize=(10, 6))

# Utiliser groupby pour tracer des lignes pour chaque nom
for nom, group in new_game2.groupby('team_name_away'):
    plt.plot(group['season'], group['somme'], marker='o', linestyle='-', label=nom)

# Ajouter des titres et des labels
plt.title('Nombre de matchs gagnÃ©s annuellement par les Ã©quipes')
plt.xlabel('AnnÃ©e')
plt.ylabel('Matchs gagnÃ©s')
plt.legend(title='Equipes')
plt.grid(True)

# Afficher le graphique
plt.show()
```
