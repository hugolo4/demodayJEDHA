```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_NBA.csv')
df
```

```python
# list = 'Phoenix Suns'
list = input("Ville suivi du nom de l'Ã©quipe : ")
df = df[df['Team'].isin([list])]
df
```

```python
# df = (df.groupby(['season', 'Team', 'win_match', 'position'], dropna=False)['overall_pick'].sum()/df.groupby(['season', 'Team', 'win_match', 'position'], dropna=False)['overall_pick'].count()).reset_index(name='total_pick')
df = df.groupby(['season', 'Team', 'win_match', 'position'], dropna=False)['overall_pick'].min().reset_index(name='total_pick')
```

```python
Max = df['total_pick'].max()
Min = 1
df['total_pick'] = df.apply(lambda x: Max + Min - x['total_pick'], axis=1)
```

```python
df
```

```python
df.fillna("NaN",inplace=True)
df
```

```python
def category(valeur):
    if valeur == 'NaN':
        pass
    elif valeur <= 10:
        return 'A'
    elif valeur <= 20:
        return "B"
    elif valeur <= 30:
        return "C"
    elif valeur <= 35:
        return "D"
    elif valeur <= 40:
        return "E"
    elif valeur <= 45:
        return "F"
    elif valeur <= 50:
        return "G"
    elif valeur <= 55:
        return "H"
    elif valeur <= 60:
        return "I"
    elif valeur <= 65:
        return "J"
    elif valeur <= 70:
        return "K"
    elif valeur <= 80:
        return "L"
    else:
        return "M"
```

```python
df['win_match'] = df['win_match'].apply(category)
df
```
